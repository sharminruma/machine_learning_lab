```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
data = {
    'Age': [25, 27, np.nan, 32, 34, 29],
    'Salary': [50000, 54000, 52000, np.nan, 62000, 58000],
    'Country': ['USA', 'Canada', 'USA', 'UK', np.nan, 'Canada']
}
df = pd.DataFrame(data)
print("---- Original DataFrame ----")
print(df)
```

```python
print("---- Initial DataFrame Info ----")
df.info()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
print("Original DataFrame:")
print(df)
# Impute missing ’Age’ with the median
age_median = df['Age'].median()
df['Age'].fillna(age_median, inplace=True)
# Impute missing ’Salary’ with the mean
salary_mean = df['Salary'].mean()
df['Salary'].fillna(salary_mean, inplace=True)
# Impute missing ’Country’ with the mode
country_mode = df['Country'].mode()[0]
df['Country'].fillna(country_mode, inplace=True)
print("\nDataFrame after Imputation:")
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Using the cleaned DataFrame from the previous step
numerical_features = df[['Age', 'Salary']]

# Min−Max Scaling
min_max_scaler = MinMaxScaler()

df_min_max_scaled = pd.DataFrame(min_max_scaler.fit_transform(numerical_features),
                                 columns=['Age_Scaled', 'Salary_Scaled'])

print("\nDataFrame after Min−Max Scaling:")
print(df_min_max_scaled)
```

```python
from sklearn.preprocessing import LabelEncoder
label_df = df.copy()
le = LabelEncoder()
label_df['Country_LabelEncoded'] = le.fit_transform(label_df['Country'])
print("Label Encoded DataFrame:\n", label_df, "\n")
```

```python
# One-Hot Encoding for the 'Country' column
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (আগের df ব্যবহার করা হয়েছে)
df_encoded = pd.get_dummies(df, columns=['Country'], prefix='Country')

# Let’s assume we have a target variable for demonstration purposes
df_temp = df_encoded.copy()
df_temp['Target'] = [1, 0, 0, 1, 1, 0]  # Dummy target

# Compute correlation matrix
correlation_matrix = df_temp.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

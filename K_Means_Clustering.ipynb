```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.cluster import KMeans
import sklearn.metrics as sm
%matplotlib inline
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
df.head()
```

```python
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
x.head()
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c='Blue')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer distribution before clustering')
plt.show()
```

```python
kmeans_model = KMeans(n_clusters=3, random_state=0)
kmeans_model.fit(x)
```

```python

labels = kmeans_model.predict(x)
colors = np.array(['gray', 'purple', 'cyan'])
cluster0 = mpatches.Patch(color='gray', label='Cluster 0')
cluster1 = mpatches.Patch(color='purple', label='Cluster 1')
cluster2 = mpatches.Patch(color='cyan', label='Cluster 2')
plt.figure(figsize=(4, 3))

plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c=colors[labels])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Clusters')
plt.legend(handles=[cluster0, cluster1, cluster2])
plt.show()
```
